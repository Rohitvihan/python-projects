```python
#GUI=Graphical User Interfaces (GUIs) are applications that allow users to interact with a
#program through visual elements like buttons, text fields, and icons instead of command-line inputs.


from tkinter import *# tkinter is used for creating window & adding widgets(buttons, labels, entry & exit boxes)
from tkinter import ttk #ttk:-used to style the widgets like background color,coloring box,font style,title
from googletrans import Translator,LANGUAGES# translator:-is used to translate the lang.LANGUAGES:-it's a library which contains
# n numb of languages


root=Tk()
root.geometry("1100x320")
root.resizable(0,0)#we cant resize the window box if it is (0,0)
root.title("language translator by Rohit Vihan K")
root["bg"]="skyblue"


Label(root,text="language translator",font="Arial 20 bold").pack()#.pack() function is used to arrange (or "pack") 
#widgets within the window in a specific layout.
Label(root,text="Entry text",font="arial 13 bold",bg="white smoke").place(x=165,y=90)#The .place() method in Tkinter allows 
#us to position widgets precisely by specifying exact coordinates within the parent window.


input_text=Entry(root,width=60)
input_text.place(x=30,y=130)
input_text.get()
#Entry: This creates an Entry widget, which is a single-line text field where users can input text.
#width=60: This specifies the width of the Entry widget, which is set to 60 characters wide.
#get(): This method is used to retrieve the text entered by the user in the Entry widget.



Label(root,text="output",font="arial 13 bold",bg="white smoke").place(x=780,y=90)
output_text=Text(root,font="arial 10",height=5,wrap=WORD,padx=5,pady=5,width=50)
output_text.place(x=600,y=180)
#height=5:-This sets the height of the Text widget in terms of the number of text lines.In this case,the height is 5 lines of text.
#wrap=WORD:-This ensures that the text wraps at word boundaries. When the text reaches the end of a line, 
#it will wrap to the next line without breaking words in half.
#padx=5:-This sets padding (space) inside the widget along the horizontal axis (left and right) to 5 pixels.
#pady=5:-This sets padding (space) inside the widget along the vertical axis (top and bottom) to 5 pixels.
#width=50:-This sets the width of the Text widget in terms of the number of characters that can fit on one line.
#Here, it allows for 50 characters per line.
#Text: This creates a Text widget in Tkinter, which allows the user to display or input multiline text. It is more flexible 
#than the Entry widget and is used for larger text areas.


language=list(LANGUAGES.values())

dest_lang=ttk.Combobox(root,values=language,width=22)
dest_lang.place(x=130,y=180)
dest_lang.set("choose languages")
#ttk.Combobox: This creates a combobox (a dropdown menu) using the ttk module of Tkinter. 
#values=language: This sets the list of values for the dropdown menu,language is presumably a list containing 
#the names of different languages (e.g., ['English', 'Spanish', 'French', ...]).

def Translate():
    translator=Translator()
    translated=translator.translate(text=input_text.get(),dest=dest_lang.get())
    output_text.insert(END, translated.text)
trans_butn=Button(root,text="Translate",font="arial 12 bold",pady=5,command=Translate,bg="orange",activebackground="green")
trans_butn.place(x=445,y=180)    
#translator=Translator():This creates an instance of the Translator class (from the googletrans library),which is used to perform the translation.
#translator.translate(...): This uses the Translator object to translate the text from the source language to the selected destination language
#output_text.insert(END, translated.text): This inserts the translated text into the output_text ( output Text widget). 
#END specifies that the text should be added at the end of any existing text in the widget.
#command=Translate:This specifies that when the button is clicked, the Translate function should be called.
#activebackground="green": This sets the background color of the button to green when the button is pressed.

root.mainloop()
```
